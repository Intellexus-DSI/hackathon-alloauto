```python
import json

# Path to your JSONL file
jsonl_path = "./results_claude-sonnet-4-5-20250929.jsonl"

total_switches = 0
sample_switch_counts = []

with open(jsonl_path, "r", encoding="utf-8") as f:
    for line in f:
        if not line.strip():
            continue
        try:
            obj = json.loads(line)
            num_preds = len(obj.get("predictions", []))
            sample_switch_counts.append((obj.get("sample_id"), num_preds))
            total_switches += num_preds
        except json.JSONDecodeError as e:
            print(f"Skipping malformed line: {e}")

print(f"Total predicted switches: {total_switches}")
print("Predicted switches per sample:")
for sid, count in sample_switch_counts:
    print(f"  Sample {sid}: {count}")
```
