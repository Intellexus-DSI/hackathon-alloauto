```python
import os
```

```python
def list_files_in_directory(directory_path):
    """Lists only the files in a single directory."""
    files_list = []
    # Check if the path exists and is a directory
    if os.path.isdir(directory_path):
        for entry in os.listdir(directory_path):
            full_path = os.path.join(directory_path, entry)
            if os.path.isfile(full_path):
                files_list.append(entry)
    return files_list
```

```python
my_dir = 'bjornpixel/Sungbum_flat_shortened'  # Current directory
sungbum_files = list_files_in_directory(my_dir)
print(len(sungbum_files))
```

```python
my_dir = 'bjornpixel/step3_output'  # Current directory
step3_files = list_files_in_directory(my_dir)
print(len(step3_files))
```

```python
sungbum_files_cut = list(map(lambda x: x[:-4], sungbum_files))
step3_files_cut = list(map(lambda x: x[:-20], step3_files))
```

```python
matching_texts = []
for f in step3_files_cut:
    if f in sungbum_files_cut:
        matching_texts.append(f)
```

```python
matching_texts = list(map(lambda x: x + ".txt", matching_texts))
```

```python
import os
import shutil

def copy_files_from_list(file_list: list, source_dir: str, dest_dir: str):
    """
    Iterates over a source directory and copies files that are present in the
    given file_list to a destination directory.

    Args:
        file_list (list): A list of filenames (not full paths) to look for.
        source_dir (str): The path to the directory to search for files.
        dest_dir (str): The path to the directory where files will be copied.
    """
    # Create the destination directory if it doesn't exist
    os.makedirs(dest_dir, exist_ok=True)
    
    # Create a set from the file_list for faster lookups
    files_to_copy = set(file_list)
    
    # Counter for files successfully copied
    copied_count = 0
    
    print(f"Searching for {len(files_to_copy)} files in '{source_dir}'...")
    
    # Iterate over all entries in the source directory
    for entry in os.listdir(source_dir):
        # Construct the full path to the source file
        source_path = os.path.join(source_dir, entry)
        
        # Check if the entry is a file and if it's in our list
        if entry in files_to_copy:
            dest_path = os.path.join(dest_dir, entry)
            try:
                # Copy the file to the destination
                shutil.copy2(source_path, dest_path)
                print(f"Copied '{entry}' to '{dest_dir}'")
                copied_count += 1
            except Exception as e:
                print(f"Error copying '{entry}': {e}")
                
    print(f"\nFinished. Successfully copied {copied_count} of {len(files_to_copy)} files.")
```

```python
copy_files_from_list(matching_texts, "bjornpixel/Sungbum_flat_shortened", "dataset")
```
